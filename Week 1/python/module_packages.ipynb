#### Module
A module is simply a Python file (.py) that contains functions, variables, or classes.

Why:
- Helps reuse code across multiple files.
- Keeps code organized and readable.
- Makes debugging and maintenance easier.

We can use Pythons built-in modules (like math, random) or create our own.


```python
# Using built-in modules
import math as m
import random

print("Square root of 25:", m.sqrt(25))
print("Value of pi:", m.pi)
print("Random number between 1 and 10:", random.randint(1, 10))
```

#### Import Specific Functions
we can import only certain parts of a module.

Why:
- Saves memory
- Improves readability (we only import what we need)

Use 'from module import function'

```python
from math import sqrt, pi

print("Square root of 16:", sqrt(16))
print("Value of pi:", pi)
```

```python
"""
Alias : giving a shorter name to a module.

why:
- Saves typing effort.
- Improves readability for long module names.

Use 'import module as alias'
"""

import numpy as np
import pandas as pd

print("NumPy array:", np.array([1, 2, 3]))
print("Pandas DataFrame:\n", pd.DataFrame({"A": [1, 2], "B": [3, 4]}))
```

```python
import math
print("Items inside math module:\n", dir(math))
```

#### Package
A package is a folder containing multiple related modules and a special file '__init__.py'.

why:
- Organize large projects.
- Create a hierarchy of code (like folders).


1 Create a folder called 'mypackage'
2 Inside it, create an empty '__init__.py'
3 Add files:
   greetings.py

__init__.py is a special Python file that tells the interpreter:

This folder is a package  not just a random directory.

```python
from mypackage.greet import gr
gr()
```

```python
from mypackage.pas import p
p("jhon")
```
