NumPy = Numerical Python, a library for fast array and matrix operations.

Why:
- Much faster than Python lists (uses C under the hood)
- Supports vectorized operations
- Widely used in ML, DL, and data analysis

**Numpy provides:**

1. extension package to Python for multi-dimensional arrays
2. closer to hardware (efficiency)
3. designed for scientific computation (convenience)
4. Also known as array oriented computing

import numpy as np

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print("NumPy Array:", arr)
print("Type:", type(arr))
```

**Why it is useful:** Memory-efficient container that provides fast numerical operations.

```python
L = range(1000)
square = []
for i in range(1000):
    if i%2==0:
        square.append(i**2)

print(square)
```

```python
#python lists
L = range(1000)
%timeit [i**2 for i in L] #[expression for item in iterable if condition]
```

```python
a = np.arange(1000)
a**2
```

```python
a = np.arange(1000)
%timeit a**2
```

#### creation

```python
a = np.array([1, 2, 3])              # From list
b = np.zeros((2, 3))                 # All zeros
c = np.ones((3, 3))                  # All ones
d = np.arange(0, 10, 2)              # Range with step
e = np.linspace(0, 1, 5)             # #start, end,equidistant point creation- number of points - 5 values between 0 and 1
f = np.random.randint(1, 10, (2, 3)) # Random integers
g = np.diag([1, 2, 3, 4]) #construct a diagonal array.
print(np.diag(g))
print("Zeros:\n", b)
print("Random:\n", f)
```

#### attributes

```python
arr = np.array([[1,2,3],[4,5,6]])

print("Shape:", arr.shape)
print("Size:", arr.size)
print("Data Type:", arr.dtype)
print("Dimensions:", arr.ndim)
```

The items of an array can be accessed and assigned to the same way as other **Python sequences (e.g. lists)**:

```python
a = np.arange(10)
print(a)
print(a[5])  #indices begin at 0, like other Python sequences (and C/C++)
```

```python
# For multidimensional arrays, indexes are tuples of integers:

a = np.diag([1, 2, 3])
print(a)
print(a[2, 1])
```

**Slicing**

```python
a = np.arange(10)

a
```

```python
a[2:9:3] # [startindex: stop : step]
```

```python
#we can also combine assignment and slicing:

a = np.arange(10)
a[5:] = 10
a
```

```python
b = np.array([4,55,788,3,900,46,43,232,4325,466,7,689,788543])
print(b[2:5:-1])
```

```python
arr = np.array([[10,20,30],[40,50,60],[70,80,90]])

print("Element [1,2]:", arr[1,2])  # row=1 col=2
print("First row:", arr[0])
print("Second column:", arr[:,1])  # all rows, col=1
print("Slice [0:2, 1:3]:\n", arr[0:2, 1:3])
```

```python
arr = np.arange(1, 10)

print("Original:", arr)
print("Reshaped (3x3):\n", arr.reshape(3,3))

a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
```

#### Operations
**All arithmetic operates elementwise**

```python
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Square:", np.square(a))
print("Mean:", np.mean(a))
print("Sum:", np.sum(a))
```

```python
b = np.ones(4) + 1

a - b
```

```python
a = np.array([1, 2, 3, 4])
b = np.array([5, 2, 2, 4])
a == b
```

```python
a > b
```

```python
#array-wise comparisions
a = np.array([1, 2, 3, 4])
b = np.array([5, 2, 2, 4])
c = np.array([1, 2, 3, 4])

np.array_equal(a, b)
```

```python
a = np.array([1, 1, 0, 0], dtype=bool)
b = np.array([1, 0, 1, 0], dtype=bool)

np.logical_or(a, b)
```
